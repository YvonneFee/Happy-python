```python
import re
import nltk
import csv
import jieba
import jieba.posseg as pseg
from jieba import analyse
from collections import Counter
```

```python
#词性标注
with open("The_three_body_problem_full/The_three_body_problem_full.txt",encoding='UTF-8')as f:
    text=f.read().encode()

text=text.decode('utf-8')

results=re.compile('[\u4e00-\u9fa5]+').findall(text)
textstr=''.join(results)

words=pseg.cut(textstr)

word_tagged=[]
for w in words:
    word_tagged.append(w)
```

```python
word_tagged
```

```python
#词性整合

pos = {}

n = -1
for (k,v) in word_tagged:
    n+=1
    if v not in pos:
        pos[v] = 1
    else:
        pos[v] += 1
print(pos)
```

```python
word_tagged_count=len(word_tagged)
word_tagged_count
```

```python
pos_sort={}

for v in pos:
    pos_sort[v]=pos[v]/word_tagged_count
```

由于老师给的网站打不开，这里选取的是人民网的国际版面进行爬取

```python
import requests
from pyquery import PyQuery as pq
url='http://world.people.com.cn/'
r = requests.get(url,verify=False)
html=r.text.encode(r.encoding).decode(r.apparent_encoding,errors='ignore')
```

```python
doc=pq(html)
lst=doc('div.headingNews div p')
lst
```

```python
for strong in lst:
    print(pq(strong).find('a').text())
```
